```python
import requests 
import pandas as pd
from bs4 import BeautifulSoup
```

```python
# link at which web data recides
link = 'https://www.pharmaceutical-technology.com/knowledge-bank/coronavirus-faqs-covid-19-categories/disease-related/'
# Getting web data
req = requests.get(link)
# parse web data
soup = BeautifulSoup(req.content, "html.parser")
```

```python
df=pd.DataFrame(columns=['Context','Answer'])
for i in range(50,99,4):
    thead = soup.find_all('div')[i]
    head=thead.find_all("button")
    tail=thead.find_all("div")[1].find_all("div")[0]
    for tr in head:
        q=tr.text.strip()
    for tr in tail:
        a=tr.strip()
    df=df.append({'Context':q,'Answer':a},ignore_index=True)
```

```python
df
```
